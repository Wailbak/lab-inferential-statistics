## Hypothesis Testing for Mean Systolic Blood Pressure

### Steps for Setting Up the Hypothesis Test

1. **Define Null and Alternative Hypotheses:**
   - **Null Hypothesis (H₀):** The mean systolic blood pressure of the group is equal to the population mean, μ₀ = 120 mm Hg.
   - **Alternative Hypothesis (H₁):** The mean systolic blood pressure of the group is not equal to the population mean, μ ≠ 120 mm Hg.

2. **Select the Level of Significance:**
   - Typically, a 5% significance level (α = 0.05) is chosen.

3. **Choose the Test Statistic:**
   - Use the t-statistic for a large sample size (n = 100).

4. **Calculate the Test Statistic:**
   - Formula: \( t = \frac{\bar{X} - μ_0}{s/\sqrt{n}} \)
   - Where \( \bar{X} = 130.1 \) mm Hg, \( s = 21.21 \) mm Hg, \( n = 100 \), \( μ_0 = 120 \) mm Hg.

5. **Decision Rule:**
   - Compare the calculated t-value against the critical t-value from the t-distribution table at α = 0.05.


### Calculation of the Test Statistic

```python
import numpy as np

# Given values
sample_mean = 130.1
population_mean = 120
sample_std_dev = 21.21
sample_size = 100

# Calculation of the test statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))
t_statistic
```

* ### **Extracting Numeric Data**

```python
numerics = breast_cancer._get_numeric_data()
numerics.head(5)
```

___
* ### **Dropping Non-Required Columns**

```python
numerics = numerics.drop(['id', 'time'], axis=1)
```

___
* ### **Standardizing the Data**

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

___
* ### **Applying PCA**

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

___
* ### **Variance Explained by Each Principal Component**

```python
variations = pca.explained_variance_ratio_ * 100
variations = np.round(variations, decimals=1)
variations
```

___
* ### **PCA Scree Plot**

```python
plot_labels = ['PC' + str(i) for i in range(1, len(variations) + 1)]

# Seting up the dark theme for the plot
sns.set(style="darkgrid")
plt.style.use("dark_background")

from matplotlib.pyplot import figure
figure(figsize=(20, 5))
plt.bar(x = plot_labels, height = variations)
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA Scree Plot')
```

___
* ### **Principal Components DataFrame**

```python
data = pd.DataFrame(pca.components_)
data = data[[0, 1, 2]]
data.columns = ['PC1', 'PC2', 'PC3']
data.head()
```
